```python
#choose the feature
X=df.iloc[:,:-1].values # all the rows and columns(except the last col)
Y=df.iloc[:,-1].values.reshape(-1,1)# the whole row and only the last col

model=Desision_Tree(min_sample_split=2,max_depth=3)# since the data is rather small
x_train,x_test,y_trian,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)

model.fit(x_train,y_trian)


```

```python
score=accuracy_score(y_test,model.predict(x_test))
print(score*100)
```
